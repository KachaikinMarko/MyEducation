# **SPECTROGRAM FUNCTION**

**Creation of spectrogram function (Dependence: time - frequency)**

```python
from scipy.signal import spectrogram
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm
from PIL import Image


def create_spectrogram_image(sample):
    """
    For a give sound sample, which is stored as an 1-d array and sample rate,
    it creates a 2d array which is a pixel by pixel information from a scipy spectrogram
    + meshcolor graph.

    It works by the following mechanism:
    1)Divides the input signal into small overlapping segments or windows.
    2)Computes the Fourier Transform of each window to obtain the frequency content.
    3)Displays the results as a 2D array or image, where the intensity of each pixel shows the amplitude of the signal at a given time and frequency.

    Obtained spectrogram image then stored as a PIL Image object as a (128,128,3)-array.
    """


    audio_data, sample_rate = sample

    f, t, Sxx = spectrogram(audio_data, sample_rate)

    Sxx_logscale = 10 * np.log10(Sxx)

    # Normalization to use a cmap so image could be in colors :)
    Sxx_normalized = (Sxx_logscale - np.min(Sxx_logscale)) / (np.max(Sxx_logscale) - np.min(Sxx_logscale))
    colormap = cm.inferno
    rgb_image = colormap(Sxx_normalized)
    # wtf is alpha channel....
    rgb_image = rgb_image[:, :, :3]

    rgb_image_uint8 = (255 * rgb_image).astype(np.uint8) #otherwise do not work


    resized_image = Image.fromarray(rgb_image_uint8).resize((128, 128)) # we want each image to be of same size
    # i added this fromarray and asarray cause i did not know how to resize all of the images in the right way :(

    return np.asarray(resized_image)


example = create_spectrogram_image(my_audio[2])


```

# **Larging the dataset functions**

**Concatenation:**

```python

def half_image_concatenation_one(image1,image2):
  """
  For each pair of images it simply takes the first of half of a first image a
  and the second half of the second image, and concatenate by the vertical axis.
  """

  if image1.shape == image2.shape:
    width = int(image1.shape[0]/2)
    return np.concatenate((image1[:,:width,:],image2[:,width:,:]),axis=1)


def half_image_concatenation_two(image1,image2):
  """
  For each pair of images it simply takes half of a first image taken from 1/4 to 3/4
  and the same hlaf of the second image, and concatenate by the vertical axis.
  """
  if image1.shape == image2.shape:
    w1= int(image1.shape[0]/4)
    w2 = 3*int(image2.shape[0]/4)
    return np.concatenate((image1[:,w1:w2,:],image2[:,w1:w2,:]),axis=1)



example1 =  create_spectrogram_image(coughvid_audio[0])
example2 =  create_spectrogram_image(coughvid_audio[1])
example3 = half_image_concatenation_one(example1,example2)
example4 = half_image_concatenation_two(example1,example2)
plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
plt.imshow(example1)
plt.title("Image 1")
plt.axis('off')
plt.subplot(2, 2, 2)
plt.imshow(example2)
plt.title("Image 2")
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(example3)
plt.title("Image 3")
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(example4)
plt.title("Image 4")
plt.axis('off')


plt.tight_layout()
plt.show()
```

# **Training loop with Adam optimizer and CrossEntropy loss**

```python
def train_model(model,data,epoch):
    model.train()
    device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
    total_losses = []
    total_accuracies = []
    for e in range(epoch):
        model.to(device)
        rate = 1e-3
        decay = 1e-1
        optimizer = torch.optim.Adam(model.parameters(), lr=rate, weight_decay=decay)
        model.train(mode=True)
        running_loss, running_acc = 0., 0
        losses = []
        accuracies = []

        inputs = torch.stack([item[0] for item in data]).to(device)
        targets = torch.tensor([item[1] for item in data]).to(device)


        outputs = model(inputs)

        loss_fn = nn.CrossEntropyLoss()
        loss = loss_fn(outputs, targets)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        optimizer.zero_grad()
        preds = torch.argmax(outputs, 1)
        running_loss += loss.item()
        losses.append(running_loss/len(data))
        running_acc += torch.sum(preds == targets)
        accuracies.append(100*running_acc/len(data))

        total_losses.append(max(losses))
        total_accuracies.append(max(accuracies))

    plt.subplot(2,1,1)
    plt.plot(range(epoch), total_losses)
    plt.ylabel('Loss')
    plt.xlabel('epoches')
    plt.subplot(2,1,2)
    plt.plot(range(epoch), total_accuracies)
    plt.ylabel('Accuracy')
    plt.xlabel('epoche')
    plt.show()


```

# **Testing the model: Accuracy, Precision-Recall,ROC**

**Accuracy**

```python
#Simplest calculation of accuracy and running loss
def test_model_accuracy(model,data):
  device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
  model.to(device)
  model.eval()
  running_loss, running_acc = 0., 0
  with torch.no_grad():
      inputs = torch.stack([item[0] for item in data]).to(device)
      targets = torch.tensor([item[1] for item in data]).to(device)

      outputs = model(inputs)

      loss_fn = nn.CrossEntropyLoss()
      loss = loss_fn(outputs, targets)
      preds = torch.argmax(outputs, 1)
      running_loss += loss.item()
      running_acc += torch.sum(preds == targets)
  return [running_loss/len(data),running_acc.item()/len(data)]
```

```python
from sklearn.metrics import roc_curve,auc

def test_model_ROC(model,data):
  device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
  model.to(device)
  model.eval()

  with torch.no_grad():
      inputs = torch.stack([item[0] for item in data]).to(device)
      targets = torch.tensor([item[1] for item in data]).to(device)

      outputs = model(inputs)

      probabilities = torch.sigmoid(outputs[:,1])
      fpr, tpr, thresholds = roc_curve(targets,probabilities)
      roc_auc = auc(fpr, tpr)
      plt.figure()

      plt.plot(fpr, tpr, color='b', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
      plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
      plt.xlabel('False Positive Rate')
      plt.ylabel('True Positive Rate')
      plt.title('Receiver Operating Characteristic (ROC) Curve')
      plt.legend(loc='lower right')
      plt.show()
```

# **Simplest CNN from scratch**

## **CNN on a small dataset**

```python
class SimpleCNN(nn.Module):
    def __init__(self):
        super(SimpleCNN, self).__init__()
        self.conv1 = nn.Conv2d(in_channels=3, out_channels=32, kernel_size=3, padding=1)
        self.pool = nn.MaxPool2d(kernel_size=2, stride=2)
        self.conv2 = nn.Conv2d(in_channels=32, out_channels=64, kernel_size=3, padding=1)
        self.conv3 = nn.Conv2d(in_channels=64, out_channels=128, kernel_size=3, padding=1)
        self.fc1 = nn.Linear(128 * 16 * 16, 512)
        self.fc2 = nn.Linear(512, 2)
        self.relu = nn.ReLU()
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        x = self.pool(self.relu(self.conv1(x)))
        x = self.pool(self.relu(self.conv2(x)))
        x = self.pool(self.relu(self.conv3(x)))
        x = x.view(-1, 128 * 16 * 16)
        x = self.relu(self.fc1(x))
        x = self.fc2(x)
        return x
```

## **Precomputing features of VGG and it's performance on larger datasets**

Starting from now, will follow the same approach as it was in the first homework: Precomputing features and speading up the code.

```python

from torchvision.models.feature_extraction import create_feature_extractor
activation = {}
def get_activation(name):
    def hook(model, input, output):
        activation[name] = output.detach()
    return hook



def precompute_model(model, dataloader):
  device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
  result_features = []
  result_targets = []
  model.to(device)
  model.features[30].register_forward_hook(get_activation('features[30]'))
  for inputs, targets in dataloader:
      inputs = inputs.to(device)
      targets = targets.to(device)
      outputs = model(inputs)
      act = activation['features[30]'].reshape(activation['features[30]'].shape[0], -1) # WE NEED TO RESHAPE TO USE LINEAR LAYER
      result_features.append(act)
      result_targets.append(targets)

  all_features = torch.cat(result_features)
  all_targets = torch.cat(result_targets)

  return torch.utils.data.TensorDataset(all_features, all_targets)


```
